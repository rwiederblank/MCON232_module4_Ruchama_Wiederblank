Assignment: Majority Element (Easy)
Problem Given a list of integers nums, return the element that appears more than n/2 times.
You may assume a majority element always exists.
Example
nums = [3, 2, 3] → 3
nums = [2, 2, 1, 1, 1, 2, 2] → 2
Goal Solve using a dictionary that maps: number -> count.


```python
def majority(nums):
  my_dict = {}
  for num in nums:
    if num not in my_dict:
      my_dict[num] = 1
    else:
      my_dict[num] = my_dict[num] + 1
  majority_of_nums = len(nums) // 2

  for key in my_dict:
    if my_dict[key] > majority_of_nums:
      return key

nums = [2, 2, 1, 1, 1, 2, 2]
print(majority(nums))
nums = [3, 2, 3]
print(majority(nums))
nums = ([1,2,3,4,4,4,4,4,4,4,5])
print(majority(nums))
```
